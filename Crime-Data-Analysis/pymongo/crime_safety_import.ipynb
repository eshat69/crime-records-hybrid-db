```python
import pandas as pd
from pymongo import MongoClient
```

```python
import pandas as pd

# 1. Load CSV
df = pd.read_csv("crime_safety_dataset.csv")
```

```python
# 2. Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["crime_safety"]         # Database name
collection = db["crime_records"] 
```

```python

# 3. Prepare data for MongoDB
docs = []
for _, row in df.iterrows():
    doc = {
        "id": int(row["id"]),
        "crime_date": pd.to_datetime(row["date"]).date().isoformat(),  # YYYY-MM-DD
        "crime_time": pd.to_datetime(row["time"]).time().isoformat(),  # HH:MM:SS
        "crime_type": row["crime_type"],
        "location": {
            "description": row["location_description"],
            "city": row["city"],
            "state": row["state"]
        },
        "victim_age": int(row["victim_age"]),
        "victim_gender": row["victim_gender"],
        "victim_race": row["victim_race"]
    }
    docs.append(doc)
```

```python
# 4. Insert into MongoDB
collection.insert_many(docs)
print(f"Inserted {len(docs)} documents into MongoDB!")
```

```python
# Example: Find all burglaries in Houston
for doc in collection.find({"crime_type": "Burglary", "location.city": "Houston"}):
    print(doc)
```
